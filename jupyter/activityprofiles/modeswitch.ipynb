```python
import pandas as pd
import csv
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
activity_file = os.path.join(os.pardir, "data", "hla_fitness-20170109.csv")
```

```python
%%time
activity_df = pd.read_csv(activity_file)
```

```python
#working activity df
wa_df = activity_df[['PID', 'ActivityDate','StepCount', 'Source']]
wa_df
```

Remove everyone with fewer than 5 activities, and whom didn't change mode

```python
activity = wa_df.groupby('PID').count()['Source'].reset_index()
pcount = len(activity)
print(str(pcount) + ' participants.')
sufficient_activity = activity[activity['Source'] >= 5]
active_pcount = len(sufficient_activity)
print(str(pcount-active_pcount) + ' participants with less than 5 active days.')
print(str(active_pcount) + ' participants remaining.')
```

```python
#sufficient activity dataframe
sa_df = wa_df[wa_df['PID'].isin(sufficient_activity['PID'])]
sa_df
```

```python
changedsource = sawa_df.groupby(['PID', 'Source']).count()['ActivityDate'].reset_index()
sourcecount = changedsource.groupby('PID').count()['Source'].reset_index()
multiplesource = sourcecount[sourcecount['Source'] > 1]
multiplesource
```

```python
#multiple source sufficient activity dataframe
mssa_df = sa_df[sa_df['PID'].isin(multiplesource['PID'])]
mssa_df.sort_values(by=['PID', 'ActivityDate'], inplace=True)
mssa_df
```

```python
mssa_participants = len(mssa_df['PID'].unique())
print(str(mssa_participants) + ' participants with more than one Source and at least 5 days of activity')
```

```python
mssa_df.reset_index(drop=True, inplace=True)
```

```python
mssa_df['sourcediff'] = mssa_df['Source'].ne(mssa_df['Source'].shift())
mssa_df['PIDdiff'] = mssa_df['PID'].ne(mssa_df['PID'].shift())
mssa_df['diff'] = mssa_df.apply(lambda x: x['PIDdiff'] or x['sourcediff'], axis=1) 
mssa_df.iloc[280:300,:]
```
