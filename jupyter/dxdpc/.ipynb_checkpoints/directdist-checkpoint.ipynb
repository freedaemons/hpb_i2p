```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from shapely.ops import nearest_points
import numpy as np
import os
import requests
```

```python
from math import radians, cos, sin, asin, sqrt
def haversine(lon1, lat1, lon2, lat2):
    try:
        """
        Calculate the great circle distance between two points 
        on the earth (specified in decimal degrees)
        """
        # convert decimal degrees to radians 
        lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
        # haversine formula 
        dlon = lon2 - lon1 
        dlat = lat2 - lat1 
        a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
        c = 2 * asin(sqrt(a)) 
        km = 6367 * c
    except TypeError:
        return ''
    return round(km,3)
```

```python
pc_latlon_file = os.path.join(os.pardir,"data","pc_latlon.csv")
pc_latlon_dd_file = os.path.join(os.pardir,"data","pc_latlon_dd.csv")
mrt_file = os.path.join(os.pardir,"data","mrt_2016.csv")

pc_latlon_df = pd.read_csv(pc_latlon_file, header=0, names=['postalcode', 'homelat', 'homelon'])
mrt_df = pd.read_csv(mrt_file)

#fixing whitespace in the homelon that prevented automatic casting to float64
pc_latlon_df['homelon'].replace(' ', '',inplace=True)
pc_latlon_df['homelon'] = pd.to_numeric(pc_latlon_df['homelon'])

print(type(pc_latlon_df['homelon'][108]))
pc_latlon_df
```

```python
#Re-assigning the crs distorts the points for some reason
crs = None
# crs = "+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs" or "+init=epsg:4326"
```

```python
mrt_geometry = [Point(xy) for xy in zip(mrt_df.LONGITUDE, mrt_df.LATITUDE)]
mrt_gdf = gpd.GeoDataFrame(mrt_df, crs=crs, geometry=mrt_geometry)
mrt_gdf
```

```python
#operational mrt stations
omrt_gdf = mrt_gdf.loc[mrt_gdf['ACTIVE'] != 0]
omrt_gdf
```

```python
home_geometry = [Point(xy) for xy in zip(pc_latlon_df.homelon, pc_latlon_df.homelat)]
home_gdf = gpd.GeoDataFrame(pc_latlon_df, crs=crs, geometry=home_geometry)
print(home_gdf)
```

```python
omrt_multipoint = omrt_gdf.geometry.unary_union
omrt_multipoint
```

```python
def nearest_mrt(point, mrts=omrt_multipoint):
    try:
        nearest = omrt_gdf.geometry == nearest_points(point,mrts)[1]
        stn_name =  omrt_gdf[nearest].STN_NAME.get_values()[0]
        stn_lat = omrt_gdf[nearest].LATITUDE.get_values()[0]
        stn_lon = omrt_gdf[nearest].LONGITUDE.get_values()[0]
    except ValueError:
        return ('', '', '')
    return (stn_name, stn_lat, stn_lon)
```

```python
%%time
pc_latlon_dd_df = home_gdf
pc_latlon_dd_df['nearest_stn'] = pc_latlon_dd_df['geometry'].apply(lambda homeloc: nearest_mrt(homeloc))
pc_latlon_dd_df['stn_lat'] = pc_latlon_dd_df['nearest_stn'].apply(lambda stn: stn[1])
pc_latlon_dd_df['stn_lon'] = pc_latlon_dd_df['nearest_stn'].apply(lambda stn: stn[2])
pc_latlon_dd_df['nearest_stn'] = pc_latlon_dd_df['nearest_stn'].apply(lambda stn: stn[0])
print(pc_latlon_dd_df)
```

```python
%%time
pc_latlon_dd_df['directdist'] = pc_latlon_dd_df.apply(lambda pc: haversine(pc['homelon'], pc['homelat'], pc['stn_lon'], pc['stn_lat']), axis=1)
print(pc_latlon_dd_df)
```

```python
desiredcolumns = ['postalcode', 'homelat', 'homelon', 'nearest_stn', 'stn_lat', 'stn_lon', 'directdist']
pc_latlon_dd_df[desiredcolumns].to_csv(pc_latlon_dd_file, index=False)
```
